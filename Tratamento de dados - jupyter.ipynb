```python
import pandas as pd
```

```python
# Trazando os dados para serem tratados
dados = pd.read_excel("base_vendas_completa.xlsx")
```

```python
# tratando a aba de clientes
clientes = pd.read_excel("base_vendas_completa.xlsx",sheet_name="clientes")
produtos = pd.read_excel("base_vendas_completa.xlsx",sheet_name="produtos")
calendario = pd.read_excel("base_vendas_completa.xlsx",sheet_name="calendario")
vendas=pd.read_excel("base_vendas_completa.xlsx",sheet_name="vendas")
```

```python
#tratando a primeira tabela "clientes"

clientes.head()
```

```python
#verificando se existe valores nulos
clientes.isnull().sum()
```

```python
# preenchendo os valores nulos da coluna idade 

idade_media = clientes["idade"].mean()
clientes["idade"] = clientes["idade"].fillna(idade_media)
clientes.head()
clientes["idade"] = clientes["idade"].astype(int)
clientes.head()
```

```python
#verificando se há duplicidade de valores na aba de clientes

clientes.duplicated().sum()
```

```python
#verificando a padronização da coluna "estado"

clientes["estado"].unique()
```

```python
# padronizando as colunas, retirando os espacos

clientes["nome"] = clientes["nome"].str.strip()
clientes["cidade"] = clientes["cidade"].str.strip()
clientes["estado"] = clientes["estado"].str.strip()
clientes.head()
```

```python
clientes["cidade"].unique()
```

```python
#verificando o tipo de dados de cada coluna

clientes.info()
```

```python
clientes.to_excel("clientes_dados_tratados.xlsx",index=False)
```

```python
# Tratando a aba "produtos"

produtos.head()
```

```python
#verificando o tipo de dados 

produtos.info()
```

```python
# verificando dados nulos

produtos.isnull().sum()
```

```python
#tratando os dados nulos com preco medio

preco_medio = produtos["preco_base"].mean()
produtos["preco_base"] = produtos["preco_base"].fillna(preco_medio)
produtos.head()
```

```python
# verificando duplicidade de dados

produtos.duplicated().sum()
```

```python
# tratando os dados inválidos de categoria e subcategoria

correcao_subcat = {
    "Mouse": "Eletrônicos",
    "Teclado": "Eletrônicos",
    "Acessórios": "Eletrônicos",
    "Camisetas": "Roupas",
    "Camiseta Azul": "Roupas",
    "Refrigerante": "Bebidas"
}

produtos["categoria"] = produtos["subcategoria"].map(correcao_subcat)
```

```python
produtos.head()
```

```python
mapa_categoria = {
    "Camisetas": "Roupas",
    "Acessórios": "Eletrônicos",
    "Bebidas": "Bebidas"
}

produtos["categoria"] = produtos["subcategoria"].map(mapa_categoria)
```

```python
produtos.head()
```

```python
correcao_nome = {
    "Camisetas": "Camiseta Azul",
    "Acessórios": "Mouse",
    "Bebidas": "Refrigerante"
}


produtos["nome_produto"] = produtos["subcategoria"].map(correcao_nome)
```

```python
produtos.head()
```

```python
produtos.isnull().sum()
```

**Não possui Valores nulos**


```python
# salvando os dados tratados na tabela "produtos"

produtos.to_excel("produtos_dados_tratados.xlsx",index=False)
```

```python
# tratando dados tabela "vendas"

vendas.head()
```

```python
#Verificando o tipo de dados

vendas.info()
```

```python
# alterando o tipo  da coluna data

vendas["data"] = pd.to_datetime(vendas["data"])
vendas.info()
```

```python
#verificando dados nulos

vendas.isnull().sum()
```

```python
# tratando os valores nulos 

vendas["desconto"] = vendas["desconto"].fillna(0)
vendas.isnull().sum()
```
